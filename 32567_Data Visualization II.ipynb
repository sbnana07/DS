### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
import warnings
warnings.filterwarnings('ignore')
```

## ----------------------------------------------------------------------------------------------------------------------------------

### Loading The Dataset

```python
df = load_dataset('Titanic')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(np.mean(df['age']))
df.drop(['deck'],axis = 1,inplace = True)
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].astype(int)
```

```python
df
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='sex', y='age', data=df ,palette= 'Set1')
plt.title('Distribution of Age with respect to Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x='sex', y='age', hue='survived', data=df ,palette= 'Set1')
plt.title('Distribution of Age with respect to Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

#### Observations:

1. The median age of female passengers who survived appears to be slightly lower than that of male passengers who survived.
2. The range of ages for male passengers who survived is wider compared to female passengers who survived.
3. There are outliers in the age distribution of both genders, especially among those who did not survive.
4. Generally, younger passengers seem to have a higher chance of survival, regardless of gender, as indicated by the lower quartile of the box plots for survivors being generally lower than for non-survivors.
5. The median age of male passengers who did not survive is slightly higher than that of female passengers who did not survive.
